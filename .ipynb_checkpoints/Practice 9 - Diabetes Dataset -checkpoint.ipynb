```python
from sklearn.datasets import load_diabetes

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
```

```python
features, labels = load_diabetes(return_X_y=True, as_frame=True)

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=1)
```

```python
dt_reg = DecisionTreeRegressor(criterion='squared_error',
                               splitter='random',
                               max_leaf_nodes=10,
                              random_state=1)

dt_reg.fit(X_train, y_train)
```

```python
#  'score' on training set.
dt_reg.score(X_train, y_train)
```

```python
#  'score' on testing set.
dt_reg.score(X_test,y_test)
```

```python
#  value of squared_error at the root node.
plt.figure(figsize=(28,8))
a = tree.plot_tree(dt_reg, filled=True)
plt.show()
```

## Hyperparameter Tuning

* Train DecisionTreeRegressor with max_depth=4 and random_state=1

-- Use criterion as ['squared_error', 'friedman_mse', 'absolute_error', 'poisson']

-- Use splitter as 'random' or 'best'

-- Use max number of features as ['auto','sqrt', 'log2'],

-- Use cost complexity pruning parameter as [0.001,0.01,0.02,0.05]

-- Use cross validation = 4

```python
from sklearn.model_selection import GridSearchCV

criterion = ['squared_error', 'friedman_mse', 'absolute_error', 'poisson']
splitter = ['random', 'best']
max_features = ['auto','sqrt', 'log2']
ccp_alpha = [0.001,0.01,0.02,0.05]

param_grid = {'criterion':criterion, 'splitter': splitter, 'max_features':max_features, 'ccp_alpha':ccp_alpha} 
dt_reg2 = DecisionTreeRegressor(max_depth =4,
                               random_state=1)
clf = GridSearchCV(dt_reg2, param_grid=param_grid, cv=4, return_train_score=True)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

The parameters differs when same model is applied on colab.

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
