# Exploring Wine Dataset with Decsion Tree Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
features, labels = load_wine(return_X_y=True, as_frame=True)

X_train, X_test, y_train, y_test = train_test_split(features,labels, test_size=0.3, random_state=1)
```

```python
dt_clf = DecisionTreeClassifier(random_state=1)
dt_clf.fit(X_train,y_train)
```

```python
dt_clf.score(X_train,y_train)
```

```python
dt_clf.score(X_test, y_test)
```

**What is the level of depth of the trained decision tree?**

```python
dt_clf.get_depth(), dt_clf.tree_.max_depth #there are two methods to get the depth
```

The depth of a tree is the maximum distance between the root and any leaf.

**Training the decision tree classifier with following properities:**

-- Use criterion as 'entropy'

-- Use splitter as 'random'; default is 'best'

-- maximum depth = 3

-- minimum samples required to split = 4

-- Use minimum impurity decrease as 0.1

-- Train the 'model' and compute the 'score' on training data and test data.

```python
dt_clf2 = DecisionTreeClassifier(criterion = 'entropy',
                                 splitter='random', 
                                 max_depth=3, 
                                 min_samples_split=4, 
                                 min_impurity_decrease=0.1,
                                random_state=1)
```

```python
dt_clf2.fit(X_train,y_train)

dt_clf2.score(X_train,y_train)
```

```python
dt_clf2.score(X_test, y_test)
```

**Entropy at the root node?**

```python
feature_names = [f'{each}' for each in features.head(0)]

plt.figure(figsize=(28,8), facecolor='w')
a = tree.plot_tree(dt_clf2, 
                   filled=True,
                  feature_names = feature_names)
plt.show()
```
